5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a 
heatmap.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Generate synthetic dataset
np.random.seed(0)
size = 100

# Base feature
x1 = np.random.normal(0, 1, size)

# Correlated features
x2 = x1 + np.random.normal(0, 0.1, size)
x3 = x1 * 2 + np.random.normal(0, 0.2, size)
x4 = np.random.normal(0, 1, size)  # not correlated

# Create DataFrame
df = pd.DataFrame({
    'Feature 1': x1,
    'Feature 2': x2,
    'Feature 3': x3,
    'Feature 4': x4
})

# Step 2: Compute correlation matrix
corr_matrix = df.corr()

# Step 3: Visualize using heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Step 1: Generate data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Step 2: Create DataFrame (optional)
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Step 3: Plot 3D surface
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Use the original meshgrid for surface plot
surf = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none', alpha=0.9)

# Add color bar
fig.colorbar(surf, shrink=0.5, aspect=10)

# Set labels and title
ax.set_title('3D Surface Plot of sin(sqrt(xÂ² + yÂ²))')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Generate data
np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Step 2: Create bubble chart
plt.figure(figsize=(10, 6))

# Scale bubble size (area), typically by squaring or multiplying for visibility
plt.scatter(df['GDP'], df['Population'], 
            s=df['Population'] * 5,  # Bubble size
            alpha=0.6, 
            c='skyblue', 
            edgecolors='black')

# Add country names as text labels
for i in range(len(df)):
    plt.text(df['GDP'][i]+20, df['Population'][i], df['Country'][i], fontsize=9)

# Labels and title
plt.title('Bubble Chart: Population vs GDP')
plt.xlabel('GDP')
plt.ylabel('Population')
plt.grid(True)
plt.tight_layout()
plt.show()
```

 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the 
markers based on the 'sizes' and 'colors' columns.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
import numpy as np

# Display plot inline (for Jupyter Notebook)
output_notebook()

# Generate random data
np.random.seed(42)
x = np.random.rand(100) * 100
y = np.random.rand(100) * 100
sizes = np.random.randint(5, 20, size=100)
colors = np.random.choice(['red', 'blue', 'green', 'orange', 'purple'], size=100)

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Create scatter plot
p = figure(title="Bokeh Scatter Plot with Size and Color", width=700, height=400)
p.scatter(x='x', y='y', size='sizes', color='colors', source=source, alpha=0.6, legend_label="Points")

# Customize plot
p.xaxis.axis_label = "X values"
p.yaxis.axis_label = "Y values"
p.legend.location = "top_left"

# Show the plot
show(p)
```

 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

 4. Create a Bokeh histogram to visualize the distribution of the given data.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Display output in notebook
output_notebook()

# Step 1: Generate data and compute histogram
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Step 2: Create the figure
p = figure(title="Histogram of Normally Distributed Data",
           x_axis_label='Value',
           y_axis_label='Frequency',
           width=700, height=400)

# Step 3: Draw histogram as quad bars
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color="skyblue", line_color="black", alpha=0.7)

# Show plot
show(p)
```

 5. Create a Bokeh heatmap using the provided dataset.
 data_heatmap = np.random.rand(10, 10)â€¨
 x = np.linspace(0, 1, 10)â€¨
 y = np.linspace(0, 1, 10)â€¨
 xx, yy = np.meshgrid(x, y)

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256

# Display output in Jupyter/Notebook
output_notebook()

# Step 1: Generate data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Step 2: Flatten for Bokeh
source = ColumnDataSource(data=dict(
    x=xx.flatten(),
    y=yy.flatten(),
    value=data_heatmap.flatten()
))

# Step 3: Create heatmap figure
p = figure(title="Bokeh Heatmap", 
           x_range=(0, 1), y_range=(0, 1), 
           width=500, height=500,
           tools="hover", tooltips=[("x", "@x"), ("y", "@y"), ("value", "@value")])

# Step 4: Draw rectangles colored by value
p.rect(x="x", y="y", width=0.1, height=0.1, source=source,
       fill_color=linear_cmap('value', Viridis256, 0, 1),
       line_color=None)

# Show the plot
show(p)
```
